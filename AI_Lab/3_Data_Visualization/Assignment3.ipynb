```python
import pandas as pd
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

iris = pd.read_csv("iris.csv") # the iris dataset is now a Pandas DataFrame
iris.head()
```

```python
# how many examples we have of each species
iris["Species"].value_counts()
```

```python
wine_reviews = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
wine_reviews.head()
```

```python
# scatter the sepal_length against the sepal_width
fig, ax = plt.subplots()

ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i], iris['SepalWidthCm'][i],color=colors[iris['Species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
sepal_length = iris['SepalLengthCm']
sepal_width = iris['SepalWidthCm']
petal_length = iris['PetalLengthCm']
petal_width = iris['PetalWidthCm']

# Create a line chart
plt.figure(figsize=(10, 6))

# Plot each line
plt.plot(sepal_length, label='Sepal Length')
plt.plot(sepal_width, label='Sepal Width')
plt.plot(petal_length, label='Petal Length')
plt.plot(petal_width, label='Petal Width')

# Set title and labels
plt.title('Measurements for Sepal Length, Sepal Width, Petal Length, and Petal Width')
plt.xlabel('Sample Index')
plt.ylabel('Measurement (cm)')

# Add legend
plt.legend()

# Display the chart
plt.show()
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = wine_reviews['points'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Wine Review Scores') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

```python
# Count the number of occurrences of each species
species_counts = iris['Species'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=['#5C5470','#BB2525','#FFF5E0'])
plt.title('Distribution of Iris Species')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate the average sepal length for each species
average_sepal_length = iris.groupby('Species')['SepalLengthCm'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
average_sepal_length.plot(kind='bar', color=['#5C5470','#BB2525','#FFF5E0'])
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Extract Sepal Length data
sepal_lengths = iris['SepalLengthCm']

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(sepal_lengths, bins=20, color='#5C5470', edgecolor='black')
plt.title('Histogram of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = iris.corr(numeric_only=True)

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

# Apply the function to the numerical columns
highlighted_table = iris.style.apply(highlight_max, subset=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

# Display the highlighted table
highlighted_table

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the wine dataset
# If not, replace 'df' with the actual variable name

# Select a feature (example: 'points') and set a target value
feature_name = 'points'
data_point = wine_reviews[feature_name].mean()  # Example: using the mean value
target_value = 90

# Define qualitative ranges
ranges = [(80, 85), (85, 90), (90, 95), (95, 100)]

# Create a bullet graph
fig, ax = plt.subplots(figsize=(8, 2))

# Plot the target line
ax.plot([target_value, target_value], [0, 1], color='black', linewidth=5)

# Plot the data point
ax.plot([data_point], [0.5], marker='o', color='blue', markersize=10)

# Add the qualitative ranges
for idx, (low, high) in enumerate(ranges):
    ax.fill_betweenx([0, 1], low, high, color=f'C{idx}', alpha=0.5)

# Set x-axis limits
ax.set_xlim(70, 100)

# Set labels and title
ax.set_yticks([])
ax.set_xlabel('Points')
ax.set_title(f'{feature_name} Bullet Graph')

# Show the bullet graph
plt.show()
```

```python
# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(iris, hue="Species") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
# One way we can extend this plot is adding a layer of individual points on top of
# it through Seaborn's striplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines
# above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
```

```python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="Species") \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
# 
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

```python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```
