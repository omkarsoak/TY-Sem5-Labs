## Basic Image Manipulation using skimage

```python
from skimage.io import imread, imshow
from skimage.transform import SimilarityTransform, warp, swirl
```

```python
#Swirling an image
image = imread("parrot.png")
swirled_img = swirl(image,rotation=0,strength=15,radius=200)
imshow(swirled_img)
```

```python
#Adding random Gaussian noise to images
from skimage.util import img_as_float, random_noise
image = img_as_float(imread("parrot.png"))
noisy = random_noise(image,var=0.1)
imshow(noisy)
```
